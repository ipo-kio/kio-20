```python
import json
with open("/home/ilya/v5.json", "r") as f:
    u = json.load(f)

print(len(u), len(u[0]), len(u[0][0]))
```

```python
T = len(u)
X = len(u[0])
Y = len(u[0][0])

h = 2 / 300
tau = 0.1
a1 = 8.2e-8 * 1e5
a2 = 8.418 * 1e5

time = T * tau
width = X * h 
height = Y * h
```

```python
def d2x(t, x, y):
    return (u[t][x + 1][y] - 2*u[t][x][y] + u[t][x - 1][y]) / (h * h)

def d2y(t, x, y):
    return (u[t][x][y + 1] - 2*u[t][x][y] + u[t][x][y - 1]) / (h * h)

def d2x1y(t, x, y):
    return d2x(t + 1, x, y) + d2y(t, x, y)

def d2xy1(t, x, y):
    return d2x(t, x, y) + d2y(t + 1, x, y)

def dt(t, x, y):
    return (u[t + 1][x][y] - u[t][x][y]) / tau
```

```python
t0 = 18
x0 = 76
y0 = 80
for a in (a1, a2):
    for d2 in (d2x1y, d2xy1):
        v = dt(t0, x0, y0) - a * d2(t0, x0, y0)
        print(0 if abs(v) < 1e-10 else v)
print(dt(t0, x0, y0))
print(dt(t0 - 1, x0, y0))
print(u[t0 - 1][x0][y0])
print(u[t0][x0][y0])
print(d2x1y(t0 - 1, x0, y0))
print(d2xy1(t0 - 1, x0, y0))
```
