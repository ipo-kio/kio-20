```python
import json
with open("/home/ilya/v3.json", "r") as f:
    u = json.load(f)

print(len(u), len(u[0]), len(u[0][0]))
```

```python
T = len(u)
X = len(u[0])
Y = len(u[0][0])

h = 2 / 300
tau = 0.1
a = 0.04

time = T * tau
width = X * h 
height = Y * h
```

```python
def d2x(t, x, y):
    return (u[t][x + 1][y] - 2*u[t][x][y] + u[t][x - 1][y]) / (h * h)

def d2y(t, x, y):
    return (u[t][x][y + 1] - 2*u[t][x][y] + u[t][x][y - 1]) / (h * h)

def d2x1y(t, x, y):
    return d2x(t + 1, x, y) + d2y(t, x, y)

def d2xy1(t, x, y):
    return d2x(t, x, y) + d2y(t + 1, x, y)

def dt(t, x, y):
    return (u[t + 1][x][y] - u[t][x][y]) / tau
```
